```python
import numpy as np
import matplotlib.pyplot as plt
import imp
gnss_codes = imp.load_source('gnss_codes', '../src/gnss_code_generation.py')
gnss_signals = imp.load_source('gnss_signals', '../src/gnss_signal_generation.py')
gnss_algorithms = imp.load_source('gnss_algorithms', '../src/gnss_algorithms.py')
gnss_filters = imp.load_source('gnss_filters', '../src/gnss_filters.py')
```

```python
filepath = '../data/g072602f.dat'
with open(filepath, 'r') as f:
    signal = np.fromfile(f, dtype=np.byte)
print('we have {0} seconds of data'.format(len(signal) / 5e6))
```

```python
fc = 1.57542e9
fi = 1.25e6
fs = 5e6
f_chip = 1.023e6
f_nav = 50.
```

```python
sv_id = 9
code = gnss_codes.gps_l1_ca(sv_id)
fd_c, n0, snr = gnss_algorithms.acquire_coarse(signal, 20e-3, 3, fs, fc, fi, code, f_chip)
print('coarse aquisition results:\n\tcoarse doppler: {0}\tsample phase (n0): {1}\tsnr: {2}'.format(fd_c, n0, snr))
tau_nav, snr = gnss_algorithms.find_nav_bit_transition(signal, fs, fc, fi, fd_c, code, f_chip, n0 / fs)
print('nav bit detection results:\n\tnavigation bit phase (bits): {0}\tsnr: {1}'.format(tau_nav * f_nav, snr))
fd_f = gnss_algorithms.acquire_fine(signal, 5e-3, 1e-3, 50, fs, fc, fi, fd_c, code, f_chip, tau_nav)
print('fine aquisition results:\nfine doppler: {0}'.format(fd_f))
```

```python
sample_index = n0
c0 = 0.
fd = fd_f + 10
phi = 0.

delay = 0.5
l_blk = 4e-3
l_sep = 4e-3
num_blk_samples = int(l_blk * fs)
t = np.arange(num_samples) / fs

b_dll = 10.  # DLL equivalent noise bandwidth
b_pll = 10.   # PLL equivalent noise bandwidth
omega_n = 5. * b_pll # natural frequency
zeta = 1.7    # damping ratio

dll_filter = gnss_filters.FirstOrderLowpass(b_dll, 1. / l_sep)
#pll_filter = gnss_filters.FirstOrderLowpass(b_pll, 1. / l_sep)
pll_filter = gnss_filters.SecondOrderLowpassV2(omega_n, zeta, 1. / l_sep)

num_blks = int(np.floor((len(signal) / fs - sample_index / fs - l_blk) / l_sep))
epochs = sample_index / fs + np.arange(num_blks + 1) * l_sep
outputs = np.zeros((9, num_blks))

for i in range(len(epochs)-1):
    
    fd_chip = f_chip * (1. + fd / fc)
    z = []
    for dly in [delay, 0., -delay]:
        code_indices = (np.floor(c0 + dly + t * fd_chip) % len(code)).astype(int)
        conj_signal_ref = (1. - 2. * code[code_indices]) * np.exp(-1j * (2 * np.pi * (fi + fd) * t + phi))
        z.append(np.sum(signal[sample_index:sample_index + num_blk_samples] * conj_signal_ref))
    early, prompt, late = z
    
    c0_error = delay * (np.abs(late) - np.abs(early)) / (np.abs(late) + np.abs(early) + 2 * np.abs(prompt))
    phi_error = np.real(prompt) * np.imag(prompt) / np.abs(prompt)**2
    
    filtered_c0_error = dll_filter.f(c0_error)
    filtered_phi_error = pll_filter.f(phi_error)
    
    # store tracking outputs
    outputs[:, i] = c0_error, phi_error, \
                    filtered_c0_error, filtered_phi_error, \
                    c0, phi, np.real(prompt), np.imag(prompt), np.abs(prompt)
    
    # update
    next_sample_index = round(epochs[i + 1] * fs)
    time_delta = (next_sample_index - sample_index) / fs
    
    phi -= filtered_phi_error
    phi += 2. * np.pi * (fi + fd) * time_delta  # add on carrier phase
    
    c0 -= filtered_c0_error
    c0 += fd_chip * time_delta
    
    sample_index = next_sample_index
```

```python
time = np.arange(num_blks) * l_sep

fig = plt.figure()

ax = fig.add_subplot(231); ax.plot(time, outputs[0,:]); ax.plot(time, outputs[2,:]);
ax = fig.add_subplot(232); ax.plot(time, outputs[1,:]); ax.plot(time, outputs[3,:]);
ax = fig.add_subplot(234); ax.plot(time, outputs[4,:] - f_chip * time);
ax = fig.add_subplot(235); ax.plot(time, outputs[5,:] - time * 2. * np.pi * (fi + fd));

ax = fig.add_subplot(233); ax.plot(time, outputs[6,:]); ax.plot(time, outputs[7,:]);
ax = fig.add_subplot(236); ax.plot(time, outputs[8,:]);

plt.show()
```

```python
now = tau_nav  # now is the time we're at in the data
c0 = 0.
fd = fd_f
phi = 0.

delay = 0.5
l_blk = 1e-3
l_sep = 1e-3
num_blks = int(np.floor((len(signal) / fs - now - l_blk) / l_sep))

# loop filter coefficients
b_dll = 10.  # DLL equivalent noise bandwidth
b_pll = 10.  # PLL equivalent noise bandwidth

omega_n = 5. * b_pll
zeta = 1.7

alpha = zeta * omega_n * l_sep
beta = omega_n**2. * l_sep**2.
gamma = 2. * alpha + beta + 4.

alpha0 = alpha + beta
alpha1 = -alpha
beta0 = 2. - alpha - beta
beta1 = alpha - 1.

b0 = (beta + 2. * alpha) / gamma
b1 = (2. * beta) / gamma
b2 = (beta - 2. * alpha) / gamma

d1 = (8. - 2. * beta) / gamma
d2 = (2. * alpha - beta - 4.) / gamma

outputs = np.zeros((9, num_blks))
filtered_c0_error = filtered_phi_error = filtered_fd_error = 0.
phi_error = last_phi_error = last_filtered_phi_error = 0.

for i in range(num_blks):
    
    # get starting sample `n0` and number of samples `ns` for block
    n0_offset, n0 = np.modf(now * fs)
    num_samples = int(l_blk * fs)
    t = np.arange(num_samples) / fs
    fd_chip = f_chip * (1. + fd / fc)
    
    # correct for quantization offset
    t_offset = n0_offset / fs
    blk_c0 = c0 - t_offset * fd_chip
    blk_phi = phi - t_offset * 2 * np.pi * (fi + fd)
    
    z = []
    for dly in [delay, 0., -delay]:
        code_indices = (np.floor(blk_c0 + dly + t * fd_chip) % len(code)).astype(int)
        conj_signal_ref = (1. - 2. * code[code_indices]) * np.exp(-1j * (2 * np.pi * (fi + fd) * t + blk_phi))
        z.append(np.sum(signal[n0:n0 + num_samples] * conj_signal_ref))
    early, prompt, late = z
    
    c0_error = delay * (np.abs(late) - np.abs(early)) / (np.abs(late) + np.abs(early) + 2 * np.abs(prompt))
    
    next_phi_error = np.real(prompt) * np.imag(prompt) / np.abs(prompt)**2
    phi_error, last_phi_error, last_last_phi_error = next_phi_error, phi_error, last_phi_error

    filtered_c0_error = (1. - .5 * l_sep * b_dll) * filtered_c0_error + .5 * l_sep * b_dll * c0_error
    
    #filtered_phi_error = (1. - 4. * l_sep * b_pll) * filtered_phi_error + 4 * l_sep * b_pll * phi_error
    #next_filtered_phi_error = alpha0 * phi_error + alpha1 * last_phi_error + beta0 * filtered_phi_error + beta1 * last_filtered_phi_error
    next_filtered_phi_error = b0 * phi_error + b1 * last_phi_error + b2 * last_last_phi_error + d1 * filtered_phi_error + d2 * last_filtered_phi_error
    filtered_phi_error, last_filtered_phi_error = next_filtered_phi_error, filtered_phi_error

    # store tracking outputs
    outputs[:, i] = c0_error, phi_error, \
                    filtered_c0_error, filtered_phi_error, \
                    c0, phi, np.real(prompt), np.imag(prompt), np.abs(prompt)
    
    # update
    phi -= filtered_phi_error
    phi += 2. * np.pi * (fi + fd) * l_sep  # add on carrier phase
    
    c0 -= filtered_c0_error
    c0 += l_sep * fd_chip
    
    now += l_sep
```

# gnss signal tracking

#### concept

The goal of tracking is to obtain estimates for our signal parameters (code phase, carrier phase, doppler frequency, carrier-to-noise ratio) over teh entire duration of the signal. For example, the entire six-hour duration of tracking a GPS satellite might look like this:

The main trend we see in in these tracking outputs is caused by satellite dynamics--that is the motion of the satellite as it rises and then sets in relation to our receiver antenna location. However, there are (relatively) smaller fluctuations in the code phase, carrier phase, and doppler frequency caused by receiver dynamics (the motion of the receiver antenna) and other phenomena such as clock errors and propagation effects of the ionosphere and neutral atmosphere.

Modeling these effects can get complicated. However, if we zoom in enough on our signal, these effects on our signal parameters become small and linear. Therefore, to track a signal only requires that we have good initial estimates of our signal parameters (code phase and doppler frequency from acquisition) and linear estimates of parameter change/error as the signal progresses.

image of thing

The method we will look at is a combined DLL/PLL tracking loop. In these tracking loops, the loop discriminator will provide parameter error/change estimates, and the loop filter will be used to smooth the effects of noise on those estimates.

#### tracking pre-requisites:

1. aquired code phase (within ~2 samples)

2. aquired doppler frequency (within ~10Hz)

#### notation

Let $t = [t_0, t_1, ..., t_i, ...]$ be the sampling time, with a zero epoch ($t_0 = 0$) as the time of the first sample in our data. Then we write a linear version of our input signal as:

$$
x(t) = AD(t)C(t - \tau)\cos((\omega_i + \omega_d)t + \phi) + \text{noise}
$$

where

$$
A = \text{signal amplitude} \\
C(t - \tau) = \text{code modulation for relevant PRN with phase offset  } \tau \\
D(t) = \text{navigation data} \\
\omega_i = \text{intermediate/center frequency} \\
\omega_d = \text{carrier Doppler frequency} \\
\phi = \text{carrier phase}
$$

#### reference generation and correlation

Let the tracking epochs be of duration $T_I$ (integration time) and with separation time $T_{sep}$. The nominal tracking scenario is $T_I = T_{sep} = 1ms$.

We generate a reference signal to correlate with our input signal for some block
$$
x_{ref}(t) = C(t - t_{start} - \hat{\tau}) \exp(j(\omega_i + \hat{\omega_d})(t - t_{start}) + \hat{\phi}))
$$

**note**: $t_{start}$ is the time of the first sample of the block for which we generate a reference, and $\hat{\tau}$, $\hat{\omega_d}$, and $\hat{\phi}$ are our signal parameter estimates ***at time $t_{start}$***.

The result of correlating our reference with our input gives us (approximately):

$$
x(t) \star x_{ref}(t) = \sum_{\text{block}} x(t) x_{ref}^*(t) \approx AR(\Delta\tau)D(t)N\text{sinc}\left(\Delta\omega_d \frac{T_I - T_s}{2}\right)\exp\left(j\left(\Delta\omega_d\frac{T_I - T_s}{2} + \Delta \phi\right)\right)
$$

#### code phase error discriminator

For the DLL, assume $\Delta\omega_d\frac{T_I - T_s}{2} << 1 \Rightarrow \text{sinc}\left(\Delta\omega_d\frac{T_I - T_s}{2}\right) \approx 1$, and ignore the phase term. Then the correlation function $R(\Delta\tau)$ between our sampled signal and our sampled code reference becomes the primary impactor of correlation magnitude.

Consider three different code phase offsets $\tau_e, \tau_p, \tau_l$, where the subscripts stand for early, prompt, and late. Let $d$ be our correlator spacing of one-half chip, $d = \frac{1}{2f_{chip}}$. We have
$$
\tau_p = \hat{\tau} \\
\tau_e = \tau_p + d \\
\tau_l = \tau_p - d
$$

If we generate references for each of these code phase offsets and correlate with our input, we get three correlator outputs.

If we have the 3 correlator outputs, how can we derive a linear error estimate for our code phase?

Let $y_e$, $y_p$, and $y_l$ be the magnitude of the early, prompt, and late correlation outputs respectively. We can always fit a polynomial of degree 2 through the three points, centered at our prompt code phase $\tau_p$. The maximum of this polynomial should yield a linear maximum likelyhood estimate for the code phase error.

Consider the points $(-d, y_e), (0, y_p), (d, y_l)$. 

We can solve $\mathbf{y} = X\mathbf{a}$ for our coefficient vector $\mathbf{a} = [a_0, a_1, a_2]^T$.

$$
\newcommand{\va}{\mathbf{a}}
\newcommand{\vy}{\mathbf{y}}
\begin{align*}
\vy &= X \va \\
\va &= X^{-1} \vy \\
X &= \left(\begin{array}{ccc}
1 & -d & d^2 \\
1 & 0 & 0 \\
1 & d & d^2 \\
\end{array}\right) \\
\end{align*}
$$

$$
\begin{align*}
|X| &= 2d^3 \\
adj(X) &= \left(\begin{array}{ccc}
0 & -d^2 & d \\
2d^3 & 0 & 2d \\
0 & d^2 & d \\
\end{array}\right)^T \\
\end{align*}
$$

$$
\begin{align*}
X^{-1} &= \frac{1}{2d^3} \left(\begin{array}{ccc}
0 & 2d^3 & 0 \\
-d^2 & 0 & d^2 \\
d & 2d & d \\
\end{array}\right) \\
\Rightarrow \\
\va = \left[ y_p, \frac{1}{2d}(y_l - y_e), \frac{1}{2d^2}(y_e + y_l + 2y_p) \right]
\end{align*} \\
$$

The next question is, where does the maximum occur for this function. It occurs where $\frac{d}{dx}\left(a_0 + a_1 x + a_2 x^2\right) = 0 \Rightarrow a_1 + 2a_2x = 0 \Rightarrow x = -\frac{a_1}{2a_2}$. The error is $-x$, so our code phase error has the full form:

$$
\tau_{err} = \frac{a_1}{2a_2} = \frac{d(y_l - y_e)}{y_e + y_l + 2y_p}
$$

#### carrier phase error discriminator

For the PLL, we assume $\Delta\omega_d\frac{T_I - T_s}{2} << \Delta \phi \text{ (which is also small) } \Rightarrow$

$$
\text{sinc}\left(\Delta\omega_d\frac{T_I - T_s}{2}\right) \approx 1 \\
\exp\left(j\left(\Delta\omega_d\frac{T_I - T_s}{2} + \Delta \phi\right)\right) \approx \exp\left(j\Delta \phi\right) \\
R(\Delta\tau) = 1
$$

Thus, the phase of the prompt correlator output gives us an estimate for the carrier phase error $\Delta \phi$, with ambiguity of $\pm \pi$ due to the unknown data bit. We would like an error estimate that is insensitive to $180^\circ$ phase transitions so that we can track through data bit changes. The Costas discriminator provides us with this property; if $Z$ is the complex correlator output, then:

$$
\phi_{err} = \frac{\text{real}(Z)\text{imag}(Z)}{|Z|^2}
$$

If $\Delta\phi = \angle Z$, then

$$
\begin{align*}
\text{real}(Z) &= \frac{1}{2}|z|D(t)\cos{\Delta\phi} \\
\text{imag}(Z) &= \frac{1}{2}|z|D(t)\sin{\Delta\phi} \\
\Rightarrow \\
\phi_{err} &= \frac{|z|^2D^2(t)\cos{\Delta\phi}\sin{\Delta\phi}}{4|z|^2} = \frac{2\sin{2\Delta\phi}}{4} \hspace{5mm} \text{ by trig. identity} \\
& \approx \Delta \phi \hspace{5mm} \text{by small angle approx.}
\end{align*}
$$

#### loop filter equations

Now we have estimates for the code phase error and carrier phase error, but they are noisy. To lessen the impact of noise on our phase estimates, we filter our error signal. We skip to the equations:

##### 1st order code delay error loop filter:

$$
\hat{L}_\tau[n] = (1 - 4 T_{sep} B) \hat{L}_\tau[n - 1] + 4 T_{sep} B L_\tau[n] \\
$$

##### 2nd order carrier phase error loop filter:

version 1:

$$
\hat{L}_\Phi[n] = \alpha_0 L_\Phi[n] + \alpha_1 L_\Phi[n - 1] + \beta_0 \hat{L_\Phi}[n - 1] + \beta_1 \hat{L_\Phi}[n - 2] \\
\alpha_0 = a_2 \omega_n T_{sep} + \omega_n^2 T_{sep}^2 \\
\alpha_1 = -a_2 \omega_n T_{sep} \\
\beta_0 = 2 - a_2 \omega_n T_{sep} - \omega_n^2 T_{sep}^2 \\
\beta_1 = a_2 \omega_n T_{sep} - 1 \\
$$

version 2:

$$
\hat{L}_\Phi[n] = \frac{1}{d_0} \left( b_0 L_\Phi[n] + b_1 L_\Phi[n - 1] + b_2 L_\Phi[n - 2] - d_1 \hat{L_\Phi}[n - 1] - d_2 \hat{L_\Phi}[n - 2] \right) \\
b_0 = \omega_n^2 T_{sep}^2 + 2 a_2 \omega_n T_{sep} \\
b_1 = 2 \omega_n^2 T_{sep}^2 \\
b_2 = \omega_n^2 T_{sep}^2 - 2 a_2 \omega_n T_{sep}\\
d_0 = \omega_n^2 T_{sep}^2 + 2 a_2 \omega_n T_{sep} + 4 \\
d_1 = 2 \omega_n^2 T_{sep}^2 - 8 \\
d_2 = \omega_n^2 T_{sep}^2 - 2 a_2 \omega_n T_{sep} + 4
$$

In my experience, the first version of the 2nd order phase error loop filter is more stable, and for the data we're tracking you can also get away with a first order filter similar to the DLL filter.

#### update step

After computing error estimates and filtered error estimates, we must do two things:

1. subtract the filtered errors from our current estimates
2. forward update our estimates to what they should be at the next tracking epoch

So


$$
\tau = \tau - \hat{L}_\tau[n]
$$

and

$$
\phi = \phi - \hat{L}_\Phi[n]
$$

Then update to the next tracking epoch. Recall that we define $T_{sep}$ as the time between the start of the current tracking epoch and the start time of the next tracking epoch (where start time refers to the sampling time of the first sample in the epoch). We do

$$
\tau = \mod_{T_{code}}\left(\tau - T_{sep}\right)
$$

and

$$
\phi = \phi + (\omega_i + \omega_d) T_{sep}
$$

***put this in a loop and we're done***

#### a quick word about code phase

The logic and notation behind code phase is a bit fuzzy. That's because, for some silly historical reason, we write the code as a function of time $C(t - \tau)$. This makes sense if we're careful about what it means. When we write $C(t)$, we imply that changes in code phase due to satellite dynamics and the other phenomena we mentioned are encapsulated in this function.

We usually treat the function $C(t)$ like it is periodic on small time scales where this approximation holds. But on longer time scales (> 200-400 ms) $C(t)$ **is not periodic** due to code compression/expansion.

Above we have looked at code phase as a time shift. This can easily be converted to sample phase $n_0$, where $C(t - \tau) = C(n/f_s - n_0/f_s)$ with the time conversion $n_0 = \tau f_s$. Its update step looks like:

$n_0 = \mod_{T_{code} * f_s} \left( n_0 - T_{sep}f_s \right)$


We can also use chip phase $c_0$. The advantage here is that it is a more sensible measure to track. Its update step looks like

$c_0 = c_0 + T_{sep} f_{d_{chip}}$
